```python
class Vehicle:
    def __init__(self,vehicle_type, make, engine, chasis_A,chasis_B , model, year, gas_consumption):
        self.vehicle_type = vehicle_type
        self.make = make
        self.engine = engine
        self.chasis_A = chasis_A
        self.model = model
        self.year = year
        self.gas_consumption = gas_consumption
        self.chasis_B =chasis_B


class Catalog:
    def __init__(self):
        self.vehicles = []

    def add_vehicle(self, vehicle):
        self.vehicles.append(vehicle)

    def display_catalog(self):
        print("Catalog of Vehicles:")
        print("====================================")
        for index, vehicle in enumerate(self.vehicles, 1):
            print(f"{index}.{vehicle.vehicle_type}-{vehicle.make} {vehicle.model} {vehicle.engine} {vehicle.year} {vehicle.chasis_A}  - {vehicle.gas_consumption}")

# Sample usage:
catalog = Catalog()

car1 = Vehicle( "carro","Toyota","2023", "Chasis A",2023,"corolla","2300FP","10g/km")
car2 = Vehicle("carro","ferrai","2022", "Chasis b",2023,"spiuder","5000FP","50g/km")
car3 = Vehicle("camioneta","mazda","2020", "Chasis A",2023,"cx-30","4000FP","20/km")
car4 = Vehicle("camion","ford","2021", "Chasis B",2019,"duster","1200FP","9/km")
car5 = Vehicle("moto","susuki","2019", "Chasis A",2019,"GRX","4000FP","20/km")

catalog.add_vehicle(car1)
catalog.add_vehicle(car2)
catalog.add_vehicle(car4)
catalog.add_vehicle(car5)


catalog.display_catalog()
```
